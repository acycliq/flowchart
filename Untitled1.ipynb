```python
import numpy as np
import pandas as pd
```

```python
def zero_class_boost(total_counts, midpoint=40, steepness=0.15):
  """
  Calculate boost for Zero class based on total gene counts.
  
  Parameters:
  - total_counts: array of total counts per cell
  - midpoint: count value where boost = 0.5 (inflection point)
  - steepness: how sharp the transition is (higher = sharper)
  
  Returns:
  - boost values between 0 and 1 (higher = more Zero class bias)
  """
  boost = 1 / (1 + np.exp((total_counts - midpoint) * steepness))
  return boost
```

```python
# - Plot the boost over a range of total_counts for one (midpoint, steepness):
import numpy as np
import matplotlib.pyplot as plt


def zero_class_boost(total_counts, midpoint=40, steepness=0.15):
    return 1 / (1 + np.exp((total_counts - midpoint) * steepness))

x = np.linspace(0, 100, 500)
y = zero_class_boost(x, midpoint=40, steepness=0.15)

plt.plot(x, y)
plt.xlabel('total_counts')
plt.ylabel('Prob')
plt.title('Zero class boost')
plt.grid(True)
plt.show()

# Compare various midpoints

# - Keep steepness fixed, vary midpoint:
x = np.linspace(0, 100, 500)
for m in [20, 40, 60, 80]:
    plt.plot(x, zero_class_boost(x, midpoint=m, steepness=0.15),
label=f'midpoint={m}')
plt.xlabel('total_counts')
plt.ylabel('Zero class prob')
plt.title('Effect of midpoint')
plt.legend()
plt.grid(True)
plt.show()




# - Keep steepness fixed, vary midpoint:
x = np.linspace(0, 100, 500)
for s in [0.05, 0.15, 0.25, 0.35, 0.45, 0.55, 0.65]:
    plt.plot(x, zero_class_boost(x, midpoint=60, steepness=s),
label=f'steepness={s}')
plt.xlabel('total_counts')
plt.ylabel('Zero class prob')
plt.title('Effect of steepness')
plt.legend()
plt.grid(True)
plt.show()
```
